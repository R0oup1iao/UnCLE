```python
import numpy as np 
import pandas as pd
import os
import tqdm

def main(name):
    data_path = f"/home/yhy/my_project/UnCLE/data/real/data/{name}"
    data = np.load(os.path.join(data_path, f"flow{name.lower()}.npz"))['data'][..., 0]
    coord = pd.read_csv(os.path.join(data_path, f"{name.lower()}_meta.csv"))[['Lng', 'Lat']].values
    os.makedirs(f"/home/yhy/my_project/UnCLE/data/real/{name}", exist_ok=True)
    np.save(f"/home/yhy/my_project/UnCLE/data/real/{name}/data_0.npy", data)
    np.save(f"/home/yhy/my_project/UnCLE/data/real/{name}/coords_0.npy", coord)
    return 0
```

```python
for name in tqdm.tqdm(["SD", "GBA", "GLA", "CA"]):
    main(name)
```
